### Grouping features
Here we can see we have education levels,marital status and Accept columns which can be grouped togather

```python
#We know that there sholdn't be more than one marital of education status for one costumer
marital_ls=[col for col in df.columns if 'marital' in str(col).strip()]
edu_ls=[col for col in df.columns if 'education' in str(col).strip()]

print(df[df[marital_ls].sum(axis=1)!=1])
print(df[df[edu_ls].sum(axis=1)!=1])
```

```python
#Define function to create a new column 

def group_col(df,list_col,new_col):
    x=2
    for col in list_col[1:]:
        df[col].replace(1,x,inplace=True)
        x+=1
        
    df[new_col]=df[list_col].sum(axis=1)

    for i in range(1,len(list_col)+1):
        df[new_col].replace(i,list_col[i-1],inplace=True)


    df.drop(columns=list_col,inplace=True)
```

```python
#Grouping Marital status and Education level

group_col(df,marital_ls,'marital_status')
group_col(df,edu_ls,'education_status')
```
